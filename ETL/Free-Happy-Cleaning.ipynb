```python
# Import dependencies
import pandas as pd
```

```python
# Load raw data files
Free_df = pd.read_csv(r"C:\Users\josep\Free-And-Happy\resources\freedom_scores.csv")
Happy_df = pd.read_csv(r"C:\Users\josep\Free-And-Happy\resources\happiness-cantril-ladder.csv")
```

```python
Happy_df
```

```python
Happy_means_df = Happy_df.groupby(['Code', 'Entity'])['Life satisfaction in Cantril Ladder (World Happiness Report 2022)'].mean().reset_index()
Happy_means_df
```

```python
Happy_means_df['happiness_rank'] = Happy_means_df['Life satisfaction in Cantril Ladder (World Happiness Report 2022)'].rank()
Happy_means_df
```

```python
Happy_means_df.rename(columns={'Life satisfaction in Cantril Ladder (World Happiness Report 2022)': 'Cantril_ladder_satisfaction_score'}, inplace=True)
Happy_means_df
```

```python
Free_means_df = Free_df.groupby('ISO_code')['pf_score','ef_score'].mean().reset_index()
Free_means_df
```

```python
Free_means_df.rename(columns={'pf_score': 'personal_freedom', 'ef_score':'economic_freedom'}, inplace=True)
Free_means_df
```

```python
Free_means_df['overall_score'] = Free_means_df.mean(axis=1)
Free_means_df
```

```python
Free_means_df.rename(columns={'ISO_code': 'Code'}, inplace=True)
Free_means_df
```

```python
Free_means_df['freedom_rank'] = Free_means_df['overall_score'].rank()
Free_means_df
```

```python
Free_Happy_df = pd.merge(Happy_means_df, Free_means_df, how="left", on="Code")
Free_Happy_df
```

```python
Free_Happy_df = Free_Happy_df.dropna()
Free_Happy_df
```

```python
corr = Free_Happy_df['Cantril_ladder_satisfaction_score'].corr(Free_Happy_df['overall_score'])
corr
```
