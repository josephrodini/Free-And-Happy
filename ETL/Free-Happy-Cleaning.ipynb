```python
# Import dependencies
import pandas as pd
```

```python
# Load raw data files
Free_df = pd.read_csv(r"C:\Users\josep\Free-And-Happy\resources\freedom_scores.csv")
Happy_df = pd.read_csv(r"C:\Users\josep\Free-And-Happy\resources\happiness-cantril-ladder.csv")
```

```python
Happy_df
```

```python
Happy_means_df = Happy_df.groupby('Code')['Life satisfaction in Cantril Ladder (World Happiness Report 2022)'].mean().reset_index()
Happy_means_df
```

```python
Happy_means_df['happiness_rank'] = Happy_means_df['Life satisfaction in Cantril Ladder (World Happiness Report 2022)'].rank()
Happy_means_df
```

```python
Happy_means_df.rename(columns={'Life satisfaction in Cantril Ladder (World Happiness Report 2022)': 'satisfaction_score'}, inplace=True)
Happy_means_df
```

```python
Free_means_df = Free_df.groupby('ISO_code')['pf_rol','ef_score','hf_score'].mean().reset_index()
Free_means_df
```

```python
Free_means_df['overall_score'] = Free_means_df.mean(axis=1)
Free_means_df
```

```python
Free_means_df.rename(columns={'ISO_code': 'Code'}, inplace=True)
Free_means_df
```

```python
Free_means_df['freedom_rank'] = Free_means_df['overall_score'].rank()
Free_means_df
```
